```python
import pandas as pd
# 덤프 파일 dump_data.csv를 판다스로 읽고 결과를 확인
```

```python
# 덤프파일 을 판다스로 읽고 결과를 확인하시오.
df = pd.read_csv('dump_data.csv')
df
```

```python
# purchase_month를 세로축으로 상품별 개수 집계를 수행하시오.
df['purchase_date'] = pd.to_datetime(df['purchase_date'])
df['purchase_month'] = pd.to_datetime(df['purchase_date']).dt.strftime("%Y%m")
res = df.pivot_table(index="purchase_month", columns="item_name",aggfunc="size",fill_value=0)
res
```

```python
# 구입 년월(세로축)별 상품의 매출 금액 합계를 집계하시오.
res = df.pivot_table(index="purchase_month",
                     columns="item_name",
                     values="item_price",
                     aggfunc="sum",
                     fill_value=0
                    )
res
```

```python
# 구입 년월 별 고객의 구입 개수를 집계하시오
res = df.pivot_table(index="고객이름",
                     columns="purchase_month",
                     values="item_name",
                     aggfunc="size",
                     fill_value=0
                    )
res
```

```python
# 구입 년월 별 지역의 판매 개수를 집계하시오.
res = df.pivot_table(index="지역",
                     columns="purchase_month",
                     values="item_name",
                     aggfunc="size",
                     fill_value=0
                    )
res
```
