```python
def getSoup(i):
    url = 'https://youtube-rank.com/board/bbs/board.php?bo_table=youtube&page={}'.format(i)
    driver.get(url)
    html = driver.page_source
    soup = BeautifulSoup(html, 'lxml')
    return soup
```

```python
for i in range(1,11):
    soup = getSoup(i)
    contents=soup.select('tr.aos-init')
    for content in contents:
        title= content.select('td.subject > a')[0].text.strip()
        category = content.select('p.category')[0].text.strip()
        subscriber = content.select('td.subscriber_cnt')[0].text
        replace_subscriber = subscriber.replace("10만", "100000")
        view = content.select('td.view_cnt')[0].text
        video = content.select('td.video_cnt')[0].text
        data = [title, category, replace_subscriber, view, video]
        results.append(data)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(results, columns =['title','category','subscriber','view','video'])
df
```

```python
df.to_excel('youtube_rank.xlsx')
```
