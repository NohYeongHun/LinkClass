## 1-1. 파일을 읽어 월별로 박스 그래프를 그리시오.

```python
f = open('seoul.csv')
data = csv.reader(f)
next(data)

month =[]
for i in range(12):
    month.append([])
for row in data:
    if row[-1] !='':
            month[int(row[0].split('-')[1])-1].append(float(row[-1])) # 최고기온.
plt.style.use('ggplot')
plt.figure(figsize=(10,5), dpi=300)
plt.boxplot(month, showfliers=False)
plt.xlabel('1~12 month')
plt.show() 

f.close()
```

## 1-2. 겨울만 월별로 박스 그래프로 그리시오.

```python
# 12~2월
f = open('seoul.csv')
data = csv.reader(f)
next(data)

month =[]
for i in range(12):
    month.append([])
for row in data:
    mon = row[0].split('-')[1] # 월 
    if row[-1] !='':
        if mon == '01':
            month[int(row[0].split('-')[1])-1].append(float(row[2]))
        if mon == '02':
            month[int(row[0].split('-')[1])-1].append(float(row[2]))
        if mon == '12':
            month[int(row[0].split('-')[1])-1].append(float(row[2]))
            
plt.style.use('ggplot')
plt.figure(figsize=(10,5), dpi=300)
plt.boxplot(month, showfliers=False) # 12, 1, 2 월
plt.xlabel('winter')
# plt.boxplot(month[0], showfliers=False)
plt.show() 

f.close()
```

## 2-(2) 상세거래파일에 가로 결합 가능한 파일은 무엇인지 설명하시오.
- 상세거래파일에 가로 결합이 가능한 파일은 상품코드 정보 및 가격이다.
- 그 이유는 상세거래파일인 df_det에는 item_id가 있고, 상품 코드 정보 및 가격 파일인 df_item에는 item_id가 있기 때문이다.
- item_id를 기준으로 df_det과 df_item은 가로 결합이 가능하다.

### - df_item, dt_det 가로로 통합
1. 상품 코드 정보 및 가격이 있는 df_item 데이터프레임을 불러온다.
2. 거래 내역이 포함된 df_det 데이터프레임을 불러온다.
3. 상품 코드에 해당하는 상품 가격과 상품의 이름을 df_det의 상품코드에 해당하는 곳에 추가해준다.
4. 해당 정보를 가지고 있는 df_item_info 데이터프레임을 만들어준다.

```python
df_item # 상품 코드 정보 및 가격.
```

```python
df_det
```

```python
df_item_info = pd.merge(
    left = df_item,
    right = df_det,
    how ='right',
    left_on='item_id',
    right_on='item_id'
    )
df_item_info
```

### - df_info, df_trade 가로로 통합
1. 고객 정보가 담겨있는 df_info 데이터프레임을 읽어온다.
2. 고객의 거래내역이 담겨있는 df_trade 데이터프레임을 읽어온다.
3. 고객 보안을 위해 세부 정보를 가져오지 말고 고객의 customer_id와 customer_name을 가져온다.
- df_custom_info에 df_info[['customer_id','customer_name']]을 가져온다.
4. df_trade에 df_custom_info의 고객이름 고객id를 고객id 기준으로 merge해서 df_trade_info 데이터프레임을 만든다.

```python
df_info
```

```python
df_trade
```

```python
df_custom_info = df_info[['customer_id','customer_name']]
df_custom_info
```

```python
df_trade_info = pd.merge(
    left = df_trade, 
    right = df_custom_info,
    how = 'left',
    left_on = 'customer_id',
    right_on = 'customer_id'
    )
df_trade_info
```
